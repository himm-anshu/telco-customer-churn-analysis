# Customer Churn Analysis

## Objective
The objective of this analysis is to identify key factors influencing customer churn and provide actionable business insights to improve customer retention.


## 1. Import Required Libraries

```python
import pandas as pd
```

## 2. Load Dataset

```python
df = pd.read_csv("Telco Customer Churn.csv")
df.head()
```

## 3. Initial Data Exploration
Understanding dataset structure, dimensions, and data types.

```python
df.shape
```

```python
df.info()
```

## 4. Data Cleaning
Handling missing values, correcting data types, and validating dataset integrity.

```python
df[df['TotalCharges'] == " "]
```

```python
df['TotalCharges'].value_counts()
```

```python
df['TotalCharges'] = df['TotalCharges'].replace(" ", pd.NA)
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(0)
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

### Cleaning Summary
- Replaced blank values in TotalCharges
- Converted TotalCharges to numeric format
- Handled missing values
- Verified no duplicate records
- Dataset is clean and ready for EDA

## 5. Exploratory Data Analysis

```python
df['customerID'].nunique()
```

### 1. Overall Churn Rate

```python
df['Churn'].value_counts()
```

```python
df['Churn'].value_counts(normalize=True) * 100
```

### Overall Churn Insight

- 26.5% of customers have churned.
- 73.5% of customers are retained.
- The dataset shows moderate class imbalance.
- Customer churn is a significant business concern and requires deeper analysis to identify key drivers.

```python
pd.crosstab(df['Contract'], df['Churn'])
```

```python
pd.crosstab(df['Contract'], df['Churn'], normalize='index') * 100
```

### Contract vs Churn Insight

- Month-to-month customers have a very high churn rate (42.7%).
- One-year contract customers show much lower churn (11.2%).
- Two-year contract customers are highly retained (only 2.8% churn).
- Longer contract duration significantly reduces churn risk.

```python
df.groupby('Churn')['tenure'].describe()
```

### Tenure vs Churn Insight

- Customers who churn have significantly lower average tenure.
- Retained customers stay much longer with the company.
- Early-stage customers are at higher risk of churn.
- Customer lifetime is strongly associated with churn behavior.

```python
pd.cut(df['tenure'], bins=[0,12,24,48,72]).value_counts()
```

```python
pd.crosstab(
    pd.cut(df['tenure'], bins=[0,12,24,48,72]),
    df['Churn'],
    normalize='index'
) * 100
```

### Tenure Category vs Churn

- Customers in the first 12 months show the highest churn rate.
- Churn rate consistently decreases as tenure increases.
- Long-term customers (48+ months) are highly retained.
- Early customer engagement is critical to reduce churn.

```python
pd.crosstab(df['TechSupport'], df['Churn'], normalize='index') * 100
```

### Tech Support Impact

- Customers without tech support have significantly higher churn.
- Customers who use tech support services are more likely to stay.
- Providing strong customer support reduces churn risk.

```python
pd.crosstab(df['InternetService'], df['Churn'], normalize='index') * 100
```

### Internet Service Impact

- Fiber optic users show higher churn compared to DSL users.
- Customers without internet service have the lowest churn.
- Service type influences customer retention.

```python
pd.crosstab(df['OnlineSecurity'], df['Churn'], normalize='index') * 100
```

### Online Security Impact

- Customers without online security are more likely to churn.
- Value-added services increase retention.

```python
df.groupby('Churn')['MonthlyCharges'].mean()
```

### Monthly Charges Insight

- Customers who churn pay higher average monthly charges.
- High pricing may increase churn risk.
- Pricing strategy should be evaluated.

```python
pd.crosstab(df['PaymentMethod'], df['Churn'], normalize='index') * 100
```

### Payment Method Impact

- Customers using electronic check have the highest churn.
- Automatic payment methods show lower churn rates.
- Encouraging auto-pay may improve retention.
